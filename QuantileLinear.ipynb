```python
from sklearn.datasets import make_regression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
```

```python
classifiers = {}
for tau in [0.1, 0.5, 0.9]:
    clf = LGBMRegressor(objective='quantile', alpha=tau)
    clf.fit(train_data[['Ticks']], train_data[['net']])
    preds = pd.DataFrame(clf.predict(test_data[['Ticks']]), columns = [str(tau)])
    classifiers[str(tau)] = {'clf': clf, 'predictions': preds}
```

```python
data = pd.DataFrame({'Ticks': test_data[['Ticks']].reset_index()['Ticks'],
              '0.1': classifiers['0.1']['predictions']['0.1'],
              '0.5': classifiers['0.5']['predictions']['0.5'],
              '0.9': classifiers['0.9']['predictions']['0.9'],
              'net': test_data[['net']].reset_index()['net']})
```

```python
data 
```

```python
fig = plt.figure();
ax = fig.add_subplot(111);
ax.plot(test_data.index, data['0.1'], color = (1,0,0), linewidth = 4, alpha = .9, label = '0.1');
ax.plot(test_data.index, data['0.5'], color = (0,0,0), linewidth = 4, alpha = .9, label = '0.5'); 
ax.plot(test_data.index, data['0.9'], color = (1,1,0), linewidth = 4, alpha = .9, label = '0.9');   
ax.plot(test_data.index, test_data['net'], linewidth = 4, alpha = .9, label = 'Actual test data'); 
plt.legend(loc="upper right")
plt.title('Quantile Regression Forecast for Test Data 2012-08-06 04:30:00 to 2013-06-30 23:30:00 ',loc='center')
plt.xlabel('Time')
plt.ylabel('kW')
plt.savefig('Quantile Test data', dpi=150);
```
